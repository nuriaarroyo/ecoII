# ***Procesos ARMA(1,1) y ARIMA(1,d,1)***

**El modelo ARMA(1,1) se define:**

$$
(1 - \phi B) X_t = (1 - \theta B) Z_t \quad ; \quad Z_t \sim \text{wn}(0, \sigma_Z^2) \tag{1}
$$

Donde si $X_t$ es estacionario, se le llama **proceso ARMA(1,1)**.  
Pero si $X_t$ no es estacionario, entonces aplicamos $\nabla^d$ a $X_t$  
hasta conseguir que sea estacionario:

$$
W_t = \nabla^d X_t
$$

En este caso, el proceso definido a continuación se conoce como  
**proceso ARIMA(1,d,1):**

$$
(1 - \phi_W B) W_t = (1 - \theta_W B) Z_t \quad ; \quad Z_t \sim \text{wn}(0, \sigma_Z^2) \tag{2}
$$


**Calculemos la función de autoco\gammaarianza y autocorrelación del proceso ARMA(1,1):**

$$
(1 - \phi B)\tilde{X}_t = (1 - \theta B)Z_t \quad ; \quad Z_t \sim \text{wn}(0, \sigma_Z^2)
$$

Donde:  
$$
\tilde{X}_t = X_t - \mu
$$

**Sin pérdida de generalidad, supongamos que** $\mu = 0$.

---

**De aquí:**

$$
X_t - \phi X_{t-1} = Z_t - \theta Z_{t-1}
$$

Despejando:

$$
X_t = \phi X_{t-1} + Z_t - \theta Z_{t-1} \tag{1}
$$

---


**Cálculo de** $\gamma_0$:

> **Ecuación (1.47):**
>
> $$
> \gamma_k = \mathbb{E}(\tilde{X}_t \tilde{X}_{t-k}) =
> \phi_1 \mathbb{E}(\tilde{X}_{t-1}\tilde{X}_{t-k}) +
> \phi_2 \mathbb{E}(\tilde{X}_{t-2}\tilde{X}_{t-k}) + \cdots +
> \phi_p \mathbb{E}(\tilde{X}_{t-p}\tilde{X}_{t-k})
> $$
> $$
> \quad + \mathbb{E}(Z_t \tilde{X}_{t-k}) -
> \theta_1 \mathbb{E}(Z_{t-1}\tilde{X}_{t-k}) - \cdots -
> \theta_q \mathbb{E}(Z_{t-q}\tilde{X}_{t-k})
> \tag{1.47}
> $$



**Note que:**

$$
X_t X_t = \phi X_{t-1} X_t + X_t Z_t - \theta X_t Z_{t-1}
$$

Entonces:

$$
\gamma _0 = \mathbb{E}(X_t X_t) = \phi \mathbb{E}(X_{t-1} X_t) + \mathbb{E}(X_t Z_t) - \theta \mathbb{E}(X_t Z_{t-1})
$$





> **Ecuación (1.48):**
>
> $$
> \mathbb{E}(Z_{t-i} \, \tilde{X}_{t-k}) = 0, \quad \text{si } k > i
> \tag{1.48}
> $$



>**Ecuación (1.51):**
>
>$$
>\mathbb{E}(Z_{t-i} \tilde{X}_t) =
>\phi_1 \mathbb{E}(Z_{t-i} \tilde{X}_{t-1}) +
>\phi_2 \mathbb{E}(Z_{t-i} \tilde{X}_{t-2}) + \cdots +
>\phi_i \mathbb{E}(Z_{t-i} \tilde{X}_{t-i}) -
>\theta_i \sigma_Z^2 \tag{1.51}
>$$
>
>con $i = 1, 2, \dots, \max(p, q)$




$$
\mathbb{E}(X_t X_t) = \phi \mathbb{E}(X_t X_{t-1}) + \mathbb{E}(Z_t^2) - \theta \mathbb{E}(Z_t Z_{t-1}) 
$$

$$
\mathbb{E}(X_t X_t) = \phi \gamma_1+ \sigma_Z^2 - \theta \mathbb{E}(Z_t Z_{t-1}) 
$$

$$
\mathbb{E}(X_t X_t) = \phi \gamma_1+ \sigma_Z^2 - \theta [ \phi  \mathbb{E}(Z_{t-1} E_{t-1}) - \theta \sigma_Z^2  ]
$$

$$
\mathbb{E}(X_t X_t) = \phi \gamma_1+ \sigma_Z^2 - \theta [ \phi  \sigma_Z^2 - \theta \sigma_Z^2  ]
$$


$$
\gamma_0 = \phi \gamma_1+ \sigma_Z^2 [ 1 - \theta (\phi - \theta )]
\tag{2}
$$


---



**Cálculo de** $\gamma_1$:

$$
\gamma_1 = \mathbb{E}(X_t X_{t-1}) =
\phi \, \mathbb{E}(X_{t-1} X_{t-2}) +
\mathbb{E}(X_{t-1} Z_t) -
\theta \, \mathbb{E}(X_{t-1} Z_{t-1})
$$

Como $\mathbb{E}(X_{t-1} Z_t) = 0$ y $\mathbb{E}(X_{t-1} Z_{t-1}) = \sigma_Z^2$, 

entonces:

$$
\gamma_1 = \phi \gamma_0 - \theta \sigma_Z^2
$$

---
**Cálculo de** $\gamma_k$:
Considerando** $k \geq 2$:

Por ejemplo, para $k = 2$:

$$
\gamma_2 = \mathbb{E}(X_t X_{t-2}) =
\phi \, \mathbb{E}(X_{t-2} X_{t-1}) -
\theta \, \mathbb{E}(Z_{t-1} X_{t-2})
$$

Pero $\mathbb{E}(Z_{t-1} X_{t-2}) = 0$, así que:

$$
\gamma_2 = \phi \gamma_1
$$

Y en general:

$$
\gamma_k = \phi \gamma_{k-1}
$$


**Resolviendo el sistema de ecuaciones al que se llega, se obtienen** $\gamma_k$ **y** $\rho_k$:

Sabemos que:

$$
\mathbb{E}(X_t) = 0
$$

Y que:

$$
\gamma_0 = \text{Var}(X_t) =
\phi \gamma_1 + \sigma_Z^2 (1 - \theta (\phi - \theta))
$$

---

**Ecuaciones recursivas para** $V_k$:

$$
\gamma_k =
\begin{cases}
\phi \gamma_0 - \theta \sigma_Z^2 & \text{si } k = 1 \\
\phi \gamma_{k-1} & \text{si } k \geq 2
\end{cases}
$$

---

**Autocorrelación teórica del proceso ARMA(1,1):**

$$
\rho_k =
\frac{\phi^{k-1} (1 - \phi\theta)(\phi - \theta)}{1 - 2\phi\theta + \theta^2},
\quad k = 1, 2, \dots
$$
