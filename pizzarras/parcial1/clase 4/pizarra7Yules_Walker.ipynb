# ***Ecuaciones de Yules-Walker***

**Las ecuaciones de Yule–Walker en su forma general**

$$
V_m = \sum_{k=1}^{p} \phi_k V_{m-k} + \sigma_Z^2 \delta_{m,0} \tag{1}
$$

donde:

- $\delta_{m,0}$ es la función delta de Kronecker:

$$
\delta_{i,j} =
\begin{cases}
1 & \text{si } i = j \\
0 & \text{si } i \ne j
\end{cases}
$$

- $p$ representa el orden del proceso autorregresivo.

---

### Ejemplo

Considere el proceso **AR(3)** siguiente:

$$
(1 - 0.1B - 0.3B^2 + 0.2B^3) X_t = Z_t \quad ; \quad \{Z_t\} \sim \text{wn}(0, \sigma_Z^2)
$$

donde:

- $\phi_1 = 0.1$
- $\phi_2 = 0.3$
- $\phi_3 = -0.2$

---

La ecuación general de Yule–Walker es:

$$
V_m = \sum_{k=1}^{3} \phi_k V_{m-k} + \sigma_Z^2 \delta_{m,0}
$$

---



### Calculemos las autocovarianzas del proceso:

Para $m = 0$:

$$
V_0 = \phi_1 V_1 + \phi_2 V_2 + \phi_3 V_3 + \sigma_Z^2 \tag{2}
$$

Para $m = 1$:

$$
V_1 = \phi_1 V_0 + \phi_2 V_1 + \phi_3 V_2 \tag{3}
$$

Para $m = 2$:

$$
V_2 = \phi_1 V_1 + \phi_2 V_0 + \phi_3 V_1 \tag{4}
$$



### Calculemos las autocovarianzas del proceso:

Para $m = 0$:

$$
V_0 = \phi_1 V_1 + \phi_2 V_2 + \phi_3 V_3 + \sigma_Z^2 ...\tag{2}
$$

Para $m = 1$:

$$
V_1 = \phi_1 V_0 + \phi_2 V_1 + \phi_3 V_2 ...\tag{3}
$$

Para $m = 2$:

$$
V_2 = \phi_1 V_1 + \phi_2 V_0 + \phi_3 V_1... \tag{4}
$$

Para $m = 3$:

$$
V_3 = \phi_1 V_2 + \phi_2 V_1 + \phi_3 V_0 ...\tag{5}
$$


A partir de las ecuaciones (3)–(5), se deduce el siguiente sistema:

$$
\begin{bmatrix}
\gamma_1 \\
\gamma_2 \\
\gamma_3
\end{bmatrix}
=
\begin{bmatrix}
\gamma_0 & \gamma_1 & \gamma_2 \\
\gamma_1 & \gamma_0 & \gamma_1 \\
\gamma_2 & \gamma_1 & \gamma_0
\end{bmatrix}
\begin{bmatrix}
\phi_1 \\
\phi_2 \\
\phi_3
\end{bmatrix}
\quad , \quad \text{donde } \phi_1 = 0.1,\ \phi_2 = 0.3,\ \phi_3 = -0.2
$$




De donde:

$$
\rho_1 = \frac{V_1}{V_0} = \frac{\phi_1 V_0 + \phi_2 V_1 + \phi_3 V_2}{V_0}
= \phi_1 + \phi_2 \rho_1 + \phi_3 \rho_2
$$


$$
\rho_2 = \frac{V_2}{V_0} = \frac{\phi_1 V_1 + \phi_2 V_0 + \phi_3 V_1}{V_0}
= \phi_1 \rho_1 + \phi_2 + \phi_3 \rho_1
$$



$$
\rho_3 = \frac{V_3}{V_0} = \frac{\phi_1 V_2 + \phi_2 V_1 + \phi_3 V_0}{V_0}
= \phi_1 \rho_2 + \phi_2 \rho_1 + \phi_3
$$



**En general**

Para cualquier proceso AR(p), las primeras $p$ autocorrelaciones se calculan a partir de las ecuaciones de Yule–Walker,  
mientras que el resto se obtiene mediante la siguiente recurrencia:

$$
\rho_k = \phi_1 \rho_{k-1} + \phi_2 \rho_{k-2} + \cdots + \phi_p \rho_{k-p}
$$


> Recordatorio:
>$$
>V_k = V_{-k}
>$$
>
---

**Sobre la estacionariedad de procesos AR(p)**

Con respecto a la estacionariedad de procesos AR(p) en general,  
se deduce a partir del **Teorema de Schur** y de los resultados de **convergencia** de las ecuaciones en diferencia de orden $p$, lo siguiente:

---

**Criterio de estacionariedad / invertibilidad**

Un proceso AR(p) o MA(q) es **estacionario** / **invertible**  
si los **módulos de los inversos de las raíces del polinomio de retraso** asociado al proceso  
son todos **menores que la unidad**.


### Ejemplo: Verificar si el proceso AR(3) mencionado antes es estacionario

$$
(1 - 0.1B - 0.3B^2 + 0.2B^3) X_t = Z_t \quad ; \quad \{Z_t\} \sim \text{wn}(0, \sigma_Z^2)
$$



**Calcule las raíces del polinomio de rezago**

$$
1 - 0.1x - 0.3x^2 + 0.2x^3 = 0
$$

- $x_1 = -1.40$
- $x_2 = 1.45 + 1.21i$
- $x_3 = 1.45 - 1.21i$



**Verificamos el módulo inverso:**

- $\left| x_1 \right| = 1.40 \Rightarrow \frac{1}{|x_1|} = \frac{1}{1.40} < 1$

- $\left| x_2 \right| = \sqrt{(1.45)^2 + (1.21)^2} \approx 1.89 \Rightarrow \frac{1}{|x_2|} \approx 0.52 < 1$





**Conclusión:**

Como las 3 raíces del polinomio de rezago tienen módulo inverso menor que 1,  
**se puede afirmar que el proceso es estacionario.**
